```python
import pickle
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
center = mpimg.imread('center.jpg')
left = mpimg.imread('left.jpg')
right = mpimg.imread('right.jpg')
plt.imshow(center)
```

```python
fig, axs = plt.subplots(1,3, figsize=(25, 25))
fig.subplots_adjust(hspace = .1, wspace=.1)
axs = axs.ravel()

axs[0].imshow(left)
axs[0].set_title('Left', fontsize=10)
axs[1].imshow(center)
axs[1].set_title('Center', fontsize=10)
axs[2].imshow(right)
axs[2].set_title('Right', fontsize=10)
```

```python
mirror = (cv2.flip(center,1))

fig, axs = plt.subplots(1,2, figsize=(25, 25))
fig.subplots_adjust(hspace = .1, wspace=.1)
axs = axs.ravel()

axs[0].imshow(center)
axs[0].set_title('Center', fontsize=10)
axs[1].imshow(mirror)
axs[1].set_title('Mirror', fontsize=10)
```
